```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import feature
from scipy import ndimage

img = cv2.imread('lenna.jpg', 0)

plt.imshow(img, cmap='gray')
plt.show()
```

```python
def plot_images(img, method):
    if method == 'Gaussian':
        imgs = [cv2.GaussianBlur(img, (0,0), sigmaX=sigma) for sigma in [1,2,4,8]]
    if method == 'Sobel':
        imgs = [ndimage.gaussian_gradient_magnitude(img, sigma) for sigma in [1,2,4,8]]
    if method == 'Laplacian':
        imgs = [ndimage.gaussian_laplace(img, sigma=sigma) for sigma in [1,2,4,8]]
    if method == 'Canny':
        imgs = [cv2.Canny(img, 80, 200, ksize, L2gradient=True) for ksize in [1,3,5,7]]

    fig, ax = plt.subplots(2, 2, figsize=(12,12))
    for (i,j) in [(x,y) for x in range(2) for y in range(2)]:
        ax[i, j].imshow(imgs.pop(0), cmap='gray')
    plt.savefig('test.png')
    plt.show()
```

```python
plot_images(img, 'Gaussian')
```

```python
plot_images(img, 'Sobel')
```

```python
plot_images(img, 'Laplacian')
```

```python
plot_images(img, 'Canny')
```
